```python
from pyspark.sql.functions import *
```

```python
start_date = "2024-01-01"
end_date = "2025-12-01"
```

```python
df = (spark.sql(f"""
          select explode(
              sequence(to_date("{start_date}"), to_date("{end_date}"), 
              interval 1 month) 
          ) as month_start_date
          """))
```

```python
df.display()
```

```python
df = (df.withColumn("date_key", date_format(col("month_start_date"), "yyyyMMdd").cast("int"))
       .withColumn("year", year(col("month_start_date")))
       .withColumn("month_name", date_format(col("month_start_date"), "yyyyMM"))
       .withColumn("month_short_name", date_format(col("month_start_date"), "MMM"))
       .withColumn("quarter", quarter(col("month_start_date")))
       .withColumn("year_quarter", concat(lit("Q"), quarter(col("month_start_date")))))
       
```

```python
df.display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("fmcg.gold.dim_date")
```
